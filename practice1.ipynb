```python
def login_req(username,password,a,b):
    if username in users and users[username]==password:
        print(a+b)
    else:
        print('UN and PW are incorrect')
```

```python
login_req('parthi','hello',a=2,b=3)
```

```python
def login_required(func):
    def wrapping(username,password,a,b):
       if username in users and users[username]==password:
        func(a,b)
       else:
            print('not authenticated')
    return wrapping
```

```python

def add(a,b):
    print(a+b)
```

```python
protected_add = login_required(add)
```

```python
protected_add('parthi','hello',1,2)
```

```python
@login_required
def add(a,b):
    print(a+b)
```

```python
add('parthi','hello',1,2)
```
